## Generating global maximum graphs (session 1-2)

```python
ket_func1 = 'ket_func1_{region}.npy'
ket_func2 = 'ket_func2_{region}.npy'
mid_func1 = 'mid_func1_{region}.npy'
mid_func2 = 'mid_func2_{region}.npy'
# now lets built it around individual global maximum
def maxVec(funcArr):
    vec = []
    for mat in funcArr:
        vec.append(np.argmax(mat))
    maxi = []
    for i, x in enumerate(vec):
        maxi.append(funcArr[i][x])
    return maxi
```

```python
reg='rACC'
ket1_max = maxVec(np.load(ket_func1.format(region=reg)))
ket2_max = maxVec(np.load(ket_func2.format(region=reg)))
mid1_max = maxVec(np.load(mid_func1.format(region=reg)))
mid2_max = maxVec(np.load(mid_func2.format(region=reg)))

plt.figure(figsize = [10,5])
labels = ['','Ket_ses1', 'Ket_ses2', 'Mid_ses1', 'Mid_ses2']
x_pos = np.arange(len(labels))
plt.boxplot([ket1_max, ket2_max, mid1_max, mid2_max])
plt.xticks(x_pos, labels)
print(f' T test for ketamine group diff {scipy.stats.ttest_rel(ket1_max, ket2_max)}')
print (f'T test for midazolam group {scipy.stats.ttest_rel(mid1_max, mid2_max)}')
```
