```python
from IntraSubjectCorr import calculateCor, createSessionMat, createSessionMatRest, calculateCorRS
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
subject_list = ['008','1307','1322','1339','1343','1351','1253','1364','1387','1390','1403','1464']
intra_corr = []
for sub in subject_list:
    df = createSessionMat(sub, '2', 18, 'leftAmg')
    mat = calculateCor(df)
    intra_corr.append(mat) # add just correlation number
    
```

```python
corMat = np.array(intra_corr)
dfISC = pd.DataFrame({'trt1_2':corMat[:,:,0][:,0], 'trt1_3':corMat[:,:,0][:,1], 'trt2_3':corMat[:,:,0][:,2]})
dfISC['scr_id'] = subject_list
import matplotlib.pyplot as plt
import seaborn as sns
f, ax = plt.subplots(1, 1)
x_col='Subject'
y_col = 'Correlation'
ax.scatter('scr_id', 'trt1_2', data= dfISC, color = "Red")
ax.scatter('scr_id', 'trt1_3', data= dfISC, color = "Blue")
ax.scatter('scr_id', 'trt2_3', data= dfISC, color = "Green")
ax.legend()
```

```python
print(f"Average correlation between script1 and 2 {np.mean(dfISC['trt1_2'])}.")
print(f"Average correlation between script1 and 3 {np.mean(dfISC['trt1_3'])}.")
print(f"Average correlation between script2 and 3 {np.mean(dfISC['trt2_3'])}.")

```
